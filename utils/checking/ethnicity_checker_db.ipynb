```python
import sqlite3
from collections import Counter
from pathlib import Path

DB_PATH = Path(
    "/Users/kaionamartinson/Desktop/Cultural-Analytics/Cultural-Analytics-AAPI/data/ethnicity_lexicon.db"
)
```

```python
def load_all_from_db():
    conn = sqlite3.connect(DB_PATH)
    cur = conn.cursor()

    noun_heads_counter = Counter()
    verb_ethnicity_dict = {}
    adj_ethnicity_dict = {}

    # noun heads
    cur.execute("SELECT noun_head, count FROM noun_heads")
    for noun, cnt in cur.fetchall():
        noun_heads_counter[noun] = cnt

    # verbs
    cur.execute("SELECT ethnicity, verb, count FROM verb_counts")
    for eth, verb, cnt in cur.fetchall():
        verb_ethnicity_dict.setdefault(eth, Counter())[verb] = cnt

    # adjectives
    cur.execute("SELECT ethnicity, adjective, count FROM adj_counts")
    for eth, adj, cnt in cur.fetchall():
        adj_ethnicity_dict.setdefault(eth, Counter())[adj] = cnt

    conn.close()
    return noun_heads_counter, verb_ethnicity_dict, adj_ethnicity_dict
```

```python
noun_heads_counter, verb_ethnicity_dict, adj_ethnicity_dict = load_all_from_db()
```
